# Data Visualisation of Original BABILong Benchmark Data
To compare our reproduced values with the original values, we create figures from the original values. We specifically use the values *from their paper* and not from the leaderboard. 

```python
import sys

from matplotlib.colors import LinearSegmentedColormap
from matplotlib.ticker import LogLocator

import seaborn as sns

import pandas as pd

sys.path.append("..")
import visualization


%matplotlib inline
%config InlineBackend.figure_format = 'pdf'
```

## Baseline Model

```python
# Load data from file
output_filename = "original-baseline-llm"
df = pd.read_csv(f"{output_filename}.csv")

accuracy_df = df
accuracy_df["correct_guess"] = accuracy_df["accuracy"].apply(lambda x: x / 100)
accuracy_df['qa'] = accuracy_df['qa'].apply(lambda x: x.upper())
accuracy_df
```

```python
visualization.create_line_chart(accuracy_df, "Original Baseline LLM Accuracy")
visualization.create_heatmap(accuracy_df, "Original Baseline LLM Accuracy")
```
