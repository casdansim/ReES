## Prompts Usage

We first load in the data from the knowlege graph from ReES' main accuracy experiment.

```python
path = Path("../extract-neo4j")
rees_df1 = pd.read_csv(path / Path("results.csv"))
rees_df2 = pd.read_csv(path / Path("results-L-50s.csv"))

rees_df = pd.concat([rees_df1, rees_df2])

rees_df = rees_df[rees_df['token_size'] <= 16]
rees_df
```

```python
# We want to group by single sample text and sum up the merged chunks
grouped_rees_df = rees_df.groupby(['token_size', 'qa', 'index'])['n_merged_chunks'].sum().reset_index()
grouped_rees_df
```

```python
grouped_rees_df = grouped_rees_df.groupby('token_size')['n_merged_chunks'].mean().reset_index()
grouped_rees_df['strategy'] = 'Superstring Elimination'
grouped_rees_df
```

We then load in the data from the knowledge graph for the GLiNER Flattening Experiment (SSE)

```python
# So this results-SSE file is the one, which has GLiNER's flattening
flattening_df1 = pd.read_csv(path / Path("results-SSE.csv"))
flattening_df2 = pd.read_csv(path / Path("results-SSE-50s.csv"))

flattening_df = pd.concat([flattening_df1, flattening_df2])
flattening_df
```

```python
# We want to group by single sample text and sum up the merged chunks
grouped_flattening_df = flattening_df.groupby(['token_size', 'qa', 'index'])['n_merged_chunks'].sum().reset_index()
grouped_flattening_df
```

```python
grouped_flattening_df = grouped_flattening_df.groupby('token_size')['n_merged_chunks'].mean().reset_index()
grouped_flattening_df['strategy'] = 'GLiNER Flat NER'
grouped_flattening_df
```

Now we merge the two dataframes

```python
concat_df = pd.concat([grouped_rees_df, grouped_flattening_df])
concat_df
```

```python
visualization.create_line_chart(concat_df,
                                experiment_title='Superstring Elimination Ablation Study',
                                y_axis='n_merged_chunks',
                                category='strategy',
                                n_types=2,
                                x_label='Token Size (Log$\mathregular{_2}$ Scale)',
                                y_label = 'Mean Prompts Used',
                                y_grid=True,
                                x_lims=(0, 18),
                                y_lims=None)
```
