```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.ticker import LogLocator
import pandas as pd
import sys

sys.path.append("..")
import visualization

%matplotlib inline
%config InlineBackend.figure_format = 'pdf'
```

```python
# Load data from file
df1 = pd.read_csv("threshold-1749033315.csv")
df2 = pd.read_csv("threshold-1749038083.csv")
df = pd.concat([df1, df2])
df
```

```python
accuracy_df = df.groupby(['threshold', 'token_size'])['correct_guess'].mean().reset_index()
accuracy_df
```

## Heatmap

```python
visualization.create_heatmap(accuracy_df, "GLiNER Threshold Accuracy", index_type='threshold', columns='token_size', values='correct_guess', x_label='Token Size', y_label='Threshold')
heatmap_data = accuracy_df.pivot(index='threshold', columns='token_size', values='correct_guess')
```

# Line Chart: Number of Entities

```python
entities_df = df.groupby(['threshold', 'token_size'])['entities'].mean().reset_index()
entities_df
```

```python
visualization.create_line_chart(
    entities_df,
    "GLiNER Threshold Entity Count",
    x_axis = 'token_size',
    y_axis = 'entities',
    category = 'threshold',
    n_types = 5,
    y_label = 'Number of Entities',
    x_lims = None,
    y_lims = None,
    y_grid = True
)
```
