## Data Visualisation of Recursive Retrieval Prompt Usage

```python
# Load RR data from files
path = Path("202506070756/")
path2 = Path("202506081353/")

df1 = pd.read_csv(path / Path("prompts-qa1.csv"))
df1["QA"] = "QA1"

df2 = pd.read_csv(path / Path("prompts-qa2.csv"))
df2["QA"] = "QA2"

df3 = pd.read_csv(path / Path("prompts-qa3.csv"))
df3["QA"] = "QA3"

df4 = pd.read_csv(path / Path("prompts-qa4.csv"))
df4["QA"] = "QA4"

df5 = pd.read_csv(path / Path("prompts-qa5.csv"))
df5["QA"] = "QA5"

df6 = pd.read_csv(path2 / Path("prompts-qa1.csv"))
df6["QA"] = "QA1"

df7 = pd.read_csv(path2 / Path("prompts-qa2.csv"))
df7["QA"] = "QA2"

df8 = pd.read_csv(path2 / Path("prompts-qa3.csv"))
df8["QA"] = "QA3"

df9 = pd.read_csv(path2 / Path("prompts-qa4.csv"))
df9["QA"] = "QA4"

df10 = pd.read_csv(path2 / Path("prompts-qa5.csv"))
df10["QA"] = "QA5"


rr_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10])
rr_df
```

```python
# Remove Answers, so we only have summaries left
rr_df = rr_df[rr_df['entity'].notna()]
rr_df
```

```python
# Group data by samples (texts and tokensize)
grouped_rr_df = rr_df.groupby(['token_size', 'index', 'QA'])['prompt'].count().reset_index()
grouped_rr_df
```

```python
# Mean data (want mean of prompts)
mean_rr_df = grouped_rr_df.groupby(['token_size', 'QA'])["prompt"].mean().reset_index()
mean_rr_df
```

```python
visualization.create_line_chart(
    df=mean_rr_df,
    experiment_title="Recursive Retrieval Prompt Usage",
    x_axis='token_size',
    y_axis='prompt',
    category='QA',
    n_types=5,
    x_label='Token Size (Log$\mathregular{_2}$ Scale)',
    y_label = 'Mean Prompts Used', 
    x_lims=(0, 18),
    y_lims=(0, 110),
    y_grid=True,
    y_grid_minor=False,
)
```

```python
mean_rr_df['Retrieval Strategy'] = 'Recursive Retrieval'
mean_rr_df
```

This visualisation is a bit out of the norm, so we do not import, but make it from the bottom.

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns_plot = sns.lineplot(
    data=concat_df,
    x='token_size',
    y='prompt',
    hue='QA',
    palette=sns.color_palette('colorblind6', n_colors=3),
    style='Retrieval Strategy',
    markers = {
        "Recursive Retrieval": 'o',
        "Basic Retrieval": 'X'
    },
    markersize=7,
    linewidth=1,
    linestyle='-',
    errorbar=None,
    dashes=False
)

experiment_title = "Recursive Retrieval Prompt Usage Ablation Study"

plt.xlim(-0.15, 18)
plt.ylim(0, 110)
plt.xscale('symlog', base = 2)

ax.grid(True, which='major', axis='x', linestyle='-', linewidth=1)
ax.xaxis.set_major_locator(LogLocator(2))
xticks_major = plt.xticks(minor=False)[0]

xticks_major = [int(xtick) for xtick in xticks_major if (float(xtick) == 0 or float(xtick) >= 1)]
xticks_major = xticks_major[:-2]
xticks_major.insert(0, 0)

# # Generate xtick labels
xtick_labels = [f"{int(tick)}k" for tick in xticks_major]
plt.xticks(xticks_major, xtick_labels, minor=False)

handles, legend = sns_plot.get_legend_handles_labels()

#Overwrite legends to have 4 spaces indentation for subcategories
indented_legend_labels = [
    'Question Type',
    '    Mean over all QAs',
    '    QA3',
    '    QA5',
    'Retrieval Strategy',
    '    Recursive Retrieval',
    '    Basic Retrieval'
]
legend = plt.legend(handles, indented_legend_labels, fontsize = 14)

for text in legend.get_texts():
    if text.get_text().strip() in ["Question Type", "Retrieval Strategy"]:  # match supergroup names
        text.set_fontweight("bold")


# Remove top and right boxes
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
ax.spines['bottom'].set_position(('outward', 10))
plt.title(experiment_title, fontsize = 20, pad=20, weight=600)
plt.xlabel('Token Size (Log$\mathregular{_2}$ Scale)', fontsize = 20, labelpad=10)
plt.ylabel('Mean Prompt Usage', fontsize = 20, labelpad=10)
# plt.legend(loc='upper center', fontsize = 16, bbox_to_anchor=(0.5, 1.14), ncols=5, frameon=False)
plt.tick_params(axis='both', labelsize = 16)
plt.tight_layout()
plt.savefig(f"{experiment_title}-linechart.pdf")
plt.show()
```

```python
rr_df = rr_df.groupby(['token_size', 'index', 'QA'])['prompt'].count().reset_index()
rr_df
```

First we run mean of means

```python
# Group by token_size, and question type
grouped_df = rr_df.groupby(['token_size', 'QA'])['prompt'].mean().reset_index()
grouped_df
```

```python
# Now take overall mean
mean_of_means = grouped_df.groupby('token_size')['prompt'].mean()
mean_of_means
```

Now we look at the overall mean

```python
overall_mean = rr_df.groupby(['token_size'])['prompt'].mean()
overall_mean
```

We conclude that mean of means and overall mean is identical here, so our earlier analysis should be correct.
